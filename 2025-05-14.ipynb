
# Daily Blog #14 – Falling (Hard) for Data Science

### May 14, 2025

I think I just unleashed the nerdiest version of myself so far this year, and I say that with full pride. I watched [FreeCodeCamp’s 6-hour Data Science lecture](https://youtu.be/ua-CiDNNj30?feature=shared), and it honestly felt like a masterclass designed exactly for me. It wasn’t just informative, it was invigorating. For the first time in a while, I felt intellectually **fed**, and not just fed—**starving for more**.

There’s so much I could unpack from the lecture, but most of my takeaways circle back to one conclusion:
**Data Science is a discipline I genuinely see myself thriving in.**
Not just surviving. Not just tolerating. **Thriving.**

Here’s why I’m falling in love with this field:

### **I. The Creative Side of Data Science**

Sure, it’s technical—there’s code, math, and a bit of system thinking involved—but it’s also *incredibly* creative. You get to explore patterns, derive insights, and transform raw facts into meaningful narratives. It’s where analytical rigor meets storytelling. That balance between logic and communication hits all the sweet spots of what I love: **clarity, structure, truth, and writing with a point.**

### **II. Transferable Tools, Versatile Applications**

The tools (Excel, SQL, Bash, Tableau, Python, R) aren’t confined to the field. They’re useful across industries, from corporate finance to biotech to research to startups. Data Science doesn’t box you in; it equips you to pivot. Learning it feels like I’m building a Swiss Army knife for the modern world.

### **III. Real-World Impact, Existential Dread Cure**

We’re living in a data-driven society, and whether it’s healthcare, climate, AI ethics, or digital infrastructure, **data scientists are shaping how we understand and improve the world.**
That thought? It's oddly soothing to my existential dread. It’s like I finally found a field that lets me do work that matters, without having to choose between impact and interest. “Those who handle the data understand the world”—and yes, I want to be one of them.

### **IV. The Math (Yes, Really)**

Thank *God* I loved math in elementary even though I used to think I hated it now because of a single bad teacher in trigonometry (10th grade flashbacks), but turns out I was just waiting for the *right* kind of math. Linear Algebra, Probability & Bayes’ Theorem, Stats, some Calculus, Big-O theory—I’m all in. Just don’t come at me with sin/cos/tan. We’re still beefing.


## **Other Finds of the Day**

### **x² = 2ˣ**

A fascinating [video by Prime Newtons](https://youtu.be/76uiFj33W5M?feature=shared) breaking down how the equation has three possible solutions:

* x = 2
* x = 4
* A non-integer solution found via **mathematical induction** using the **Lambert W function**, which I found sexy (yes, math can be hot):
  [Lambert W function resource](https://mathworld.wolfram.com/LambertW-Function.html)

### **Simpson's Paradox**

A classic statistical phenomenon where trends appear in different groups of data but disappear or reverse when the groups are combined. Reality check: *Data can lie.* Interpretation is everything.
