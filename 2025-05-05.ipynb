##### What I Learned
##### Data Communications & Networking Quick Summary

**Core Concepts**
* **Multicast**: Sends data to multiple specific recipients (not all), more efficient than broadcast.
* **Cloud Storage**: Storing personal files over the Internet (e.g., Google Drive).
* **QoS (Quality of Service)**: Prioritizes time-sensitive traffic (VoIP, video).
* **Network Protocols**: Define communication rules (format, error checking, addressing).
* **Open Standards**: Enable multi-vendor compatibility, innovation, flexibility.

**OSI Layers (Know these!)**
1. **Physical** – Cables, signals, bits
2. **Data Link** – MAC address, framing, error detection
3. **Network** – IP, routing
4. **Transport** – TCP/UDP, segmentation, reliability
5. **Session** – Connections, sessions
6. **Presentation** – Encryption, formatting
7. **Application** – Interface for users (HTTP, DNS)

**Network Characteristics**
* **Fault Tolerance**
* **Scalability**
* **QoS**
* **Security**

**Network Types**
* **LAN** – local area (office, home)
* **WAN** – large area (Internet)
* **MAN** – city-wide
* **PAN** – Bluetooth, personal range

**Communication Rules & Media**
* **Data Link Rules**: MAC addressing, framing, flow control
* **Media Types**:
  * **Guided**: copper, fiber
  * **Unguided**: wireless
* **Fiber-Optic**: high speed, EMI-resistant, long range, expensive
* **Bandwidth**: Maximum data transfer capacity

**Switch Features**
* **MAC Table** – stores device MAC addresses
* **Forwarding** – sends frame to correct port
* **Frame Forwarding Types**:
  * Store-and-Forward (checks for errors)
  * Cut-Through (faster, less reliable)
  * Fragment-Free (checks 64 bytes)

**Cables**
* **Straight-through** – PC ↔ switch
* **Crossover** – PC ↔ PC or switch ↔ switch
* **Rollover** – console cable

**Signal Loss & Management**
* **Attenuation** – signal weakening over distance
* **Remote Management**: via **Switch Virtual Interface (SVI)**

**SSH vs Telnet**
* **SSH** – encrypted, secure
* **Telnet** – plaintext, insecure

**IP Protocol Characteristics**
* **Connectionless**
* **Best-effort delivery**
* **Logical addressing**
* **Used for routing**

**Common Troubleshooting Tips**
* Interface down? → Use `no shutdown`
* Wrong gateway/IP? → No outside network access
* No DHCP? → No IP assigned to device

**Cisco IOS Modes & Prompts**
| Mode                        | Prompt                 | Purpose                                               |
| --------------------------- | ---------------------- | ----------------------------------------------------- |
| **User EXEC**               | `Switch>`              | Basic view-only access (ping, show)                   |
| **Privileged EXEC**         | `Switch#`              | Full access to view configurations and files          |
| **Global Configuration**    | `Switch(config)#`      | Configure system-wide settings                        |
| **Interface Configuration** | `Switch(config-if)#`   | Configure specific interfaces (e.g., FastEthernet0/1) |
| **Line Configuration**      | `Switch(config-line)#` | Configure console, SSH, Telnet access                 |
| **VLAN Interface Config**   | `Switch(config-if)#`   | Assign IP to virtual interface (e.g., VLAN 1)         |

**Common Configuration Commands**

Basic Access & Setup

```bash
enable                      # Enter privileged EXEC mode
configure terminal          # Enter global config mode
hostname <name>             # Set device name
no shutdown                 # Enable interface
```

Passwords and Access
```bash
enable secret <password>    # Encrypted privileged EXEC password
line console 0              # Access console line config
password <pw>               # Set console password
login                       # Enable password prompt
line vty 0 4                # Access SSH/Telnet config
service password-encryption # Encrypt plain text passwords
```

IP & VLAN Configuration
```bash
interface vlan 1            # Enter VLAN 1 (SVI)
ip address 192.168.1.1 255.255.255.0  # Assign IP address
no shutdown                 # Bring VLAN interface up
```

Banner Configuration
```bash
banner motd # Unauthorized access is prohibited #  # Set message of the day
```

Save & Exit
```bash
copy running-config startup-config  # Save config
end                                 # Exit to privileged mode
exit                                # Exit current mode
```

Testing Connectivity
```bash
ping <IP address>           # Test connectivity
show ip interface brief     # View IP/interface status
show mac address-table      # View learned MACs
```


Quick Mode Navigation Tip:
* From **User EXEC (`>`)**:
  `enable` → Privileged EXEC (`#`)
* From **Privileged EXEC (`#`)**:
  `configure terminal` → Global Config (`(config)#`)
* From **Global Config (`(config)#`)**:
  `interface vlan 1` → Interface Config (`(config-if)#`)
  `line console 0` or `line vty 0 4` → Line Config (`(config-line)#`)
