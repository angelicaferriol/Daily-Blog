# Daily Blog #26 – Cognitive Load Theory 
### May 26, 2025

---

### What is Cognitive Load Theory (CLT)?

Cognitive Load Theory is about how your brain’s **working memory** has **limited capacity**. Think of it like your RAM — if you overload it, performance tanks.

There are 3 types of cognitive load:

1. **Intrinsic Load** – The inherent difficulty of the material.
2. **Extraneous Load** – Unnecessary complexity in how it’s presented.
3. **Germane Load** – The effort used to create meaning and build long-term knowledge (schemas).

Your job as a learner? **Minimize extraneous load**, **optimize germane load**, and **pace intrinsic load** based on your current level.

---

### Why this matters for you?

You’re likely overloading yourself with:

* Too many open tabs
* Dense PDF readings
* Watching 2x speed videos while taking notes
* Switching topics mid-study

You’re burning **RAM** without storing **anything** in long-term memory.

---

### How to apply it *strategically*?

#### 1. **Chunk information**

Don’t memorize an algorithm line-by-line. Group concepts into patterns or processes.

> Example: Divide-and-conquer = “Split → Solve recursively → Merge.”

#### 2. **Dual-coding**

Combine **visuals + text**. Don’t just read code—draw flowcharts.

> Sketch quick diagrams of how recursion splits input.

#### 3. **Eliminate noise**

Kill all distractions and badly designed learning materials.

> If the prof’s slides suck, recreate them *your way*.

#### 4. **Use worked examples, then fade**

Study solved problems first. Then slowly try problems with less scaffolding.

> Don’t brute-force problems right away. Understand first, apply second.

#### 5. **Space your learning**

Cramming overloads working memory. Spaced repetition is how you build durable schemas.

---

### Bottom Line

If you’re feeling stuck or overwhelmed when learning algorithms or any complex topic, it’s *not just* a motivation issue. You’re probably violating Cognitive Load Theory and flooding your brain with unstructured inputs.

Fix your cognitive architecture first — that’s how high-performers *scale*.
