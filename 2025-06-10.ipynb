# Daily Blog #41 - Data Transformation in Spreadsheets
### June 10, 2025


#### **What is Data Transformation?**

- Data transformation is when you **change, restructure, or reformat data** so it’s easier to work with or analyze.
- It’s a core part of **cleaning and prepping data** for deeper analysis.



**Common Data Transformation Techniques in Excel**

**Splitting Columns**

* Break one column into multiple.
* Example: Split “First Last” into two columns (“First” and “Last”).

**Combining Columns**

* Merge data from multiple columns.
* Example: Create “Full Name” by merging “First” and “Last.”
* Formula:

  ```excel
  =A1 & " " & B1
  ```

  or

  ```excel
  =CONCATENATE(A1, " ", B1)
  ```

**Text-to-Columns**

* Use Excel’s built-in **“Data > Text to Columns”** to split data based on a delimiter (like a comma, tab, or space).

**Using Formulas to Transform Data**

* Change the **case** of text:

  * `=UPPER()`, `=LOWER()`, `=PROPER()`
* Extract parts of a string:

  * `=LEFT()`, `=RIGHT()`, `=MID()`

**Reformatting Dates**

* Dates can be messy!
* Change “MM/DD/YYYY” to “YYYY-MM-DD” using `TEXT()` function:

  ```excel
  =TEXT(A1, "yyyy-mm-dd")
  ```
