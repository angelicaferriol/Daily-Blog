# Daily Blog #9 – Rest or Escapism
### May 9, 2025

##### Why This Entry Matters  
I honestly did nothing today except eating, sleeping, and holding my phone. A part of my brain had won the argument--that I deserve to do nothing today because the previous days were stressful. My cortisol level dropped intensely after the exams, and oddly my body craved food so much to the extent that I ate several times. I also had a hard time falling asleep and relaxing my body. Today, I failed to setup a recovery plan post-exam.

##### Find of the Day
I just experienced another hypnic jerk again (it happens quite often, say thrice a week or worse once a day). Hypnic Jerk (aka Sleep Start) is a sudden, involuntary muscle twitch that often happens just as you're about to fall asleep, usually accompanied by a feeling of falling or being jolted awake. That which I initially thought I was experiencing but after sharing my symptoms to ChatGPT, it mentioned some few interesting takes:
1. **Hypervigilance / Mental Overdrive**: If your mind is high-strung, especially after stress (like exams), your body may resist sleep even when you’re tired. It’s a nervous system glitch—your brain interprets the loss of control as a "threat" and reacts like you were caught off guard.
2.  **Nervous System Dysregulation**: Common with burnout, caffeine overload, or sleep deprivation. The sympathetic nervous system (fight-or-flight) is still active when the parasympathetic (rest-and-digest) should take over.
3. **Conditioned Response**: If you’ve pushed yourself to stay awake through exhaustion repeatedly, your body may have built a "sleep = danger" association—so the moment you start drifting, your body fires off an alert.

##### Next Steps as Generated by AI

**1. Physical Reset (2–3 Days)**
**Objective:** Calm your overstimulated nervous system and repair your body.
* *Sleep Debt Recovery*: Prioritize 8–10 hours/night for 2–3 days. No alarms.
* *Real Meals*: Reintroduce protein-rich, fiber-filled, slow-carb food. Hydrate.
* *Do-nothing Blocks*: Schedule intentional “boredom time” without guilt. Let your mind wander—this rebuilds creative thinking.
* *Optional challenge:* 24h digital detox.

**2. Mental Declutter & Reorientation (1–2 Days)**
**Objective:** Reclaim clarity and purpose. Burnout fog = mental chaos.
* *Dump your brain*: Journal or mindmap all the “unfinished” thoughts, stress, and guilt still bouncing around.
* *Do a Weekly Reset (Notion, planner, whatever)*: Clear old tasks, resync calendars, and **see what actually matters now**.
* *Set “no pressure” intentions*: Not goals, just **what kind of energy you want to build next week** (e.g., exploratory, restorative, focused).

**3. Curiosity-Based Re-engagement (3–5 Days)**
**Objective:** Reignite passion without pressure. Burnout thrives on drudgery.
* Pick **one low-stakes skill or hobby** you’ve ignored (digital painting, short story, Notion template, violin prep).
* Give it **1 hour/day max**, just to enjoy doing something *you choose*, not assigned.
* Let progress feel optional. This is about freedom.
