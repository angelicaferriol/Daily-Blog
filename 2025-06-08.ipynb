# Daily Blog #39 -  Key Excel Skills for Data Work
### June 8, 2025

### **1. Data Cleaning: Filters and Conditional Formatting**

**Filters**:
* Let you **quickly isolate** rows matching criteria.
* Example: Filter all rows where “Status” is “Inactive” to see how many inactive sites you have.

**Conditional Formatting**:
* Adds color-based rules → instant pattern spotting.
* Examples:

  * Highlight cells with missing data (`=ISBLANK` logic).
  * Highlight cells > 100 if you’re tracking large site areas.
  * Color duplicate entries in site codes so you can fix them fast.


### **2. Data Validation: Control Inputs**
**Set allowed values** for a column (dropdown lists, numeric limits).
* Example: In a marine sites table, force “Region” to only be: “NCR”, “CAR”, “Region IV-B”.
* Prevents messy data entry like “reg iv b” or “ncrr.”


### **3. Pivot Tables: Summarize Massive Data**

**How it works**:
* Drag-and-drop columns to rows/values to create **summary tables** in seconds.
* Example:
  * You have data of **reef areas** across 30 islands.
  * Create a Pivot Table: Rows = “Island Name”, Values = “SUM of reef area” or “Average coral cover.”


### **4. VLOOKUP / XLOOKUP: Merge Datasets**

**VLOOKUP** syntax:

```excel
=VLOOKUP(lookup_value, table_array, col_index, [range_lookup])
```

* Example:

  * **Sheet 1**: Reef sites with site ID.
  * **Sheet 2**: Site ID + region info.
  * **Goal**: Pull region info into Sheet 1 using `VLOOKUP(siteID, Sheet2, 2, FALSE)`.

**XLOOKUP** is better (introduced in newer Excel versions):
* Works left-to-right **and** right-to-left (VLOOKUP only does left-to-right).
* Example:

```excel
=XLOOKUP(siteID, Sheet2!A:A, Sheet2!B:B)
```

### **5. Named Ranges & Dynamic Charts: Clean, Scalable Analysis**

**Named Ranges**:
* Instead of `=A1:A100`, use `=site_data`.
* Example: Name `A1:A100` as `reef_sites` → easier to manage formulas.

**Dynamic Charts**:
* Charts that auto-update when data grows.
* Use “Tables” (Insert → Table) + dynamic named ranges for this.
