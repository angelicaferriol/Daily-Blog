# Daily Blog #44 - Merge Sort
### June 13, 2025

### Merge Sort Pseudocode

```plaintext
MergeSort(arr, start, end)
    if start < end
        mid = (start + end) / 2
        MergeSort(arr, start, mid)
        MergeSort(arr, mid + 1, end)
        Merge(arr, start, mid, end)
```


### Merge Function Pseudocode

```plaintext
Merge(arr, start, mid, end)
    n1 = mid - start + 1
    n2 = end - mid

    // Create temp arrays
    left[1..n1]
    right[1..n2]

    // Copy data to temp arrays
    for i = 0 to n1 - 1
        left[i] = arr[start + i]

    for j = 0 to n2 - 1
        right[j] = arr[mid + 1 + j]

    // Merge the temp arrays back into arr[start..end]
    i = 0, j = 0, k = start
    while i < n1 and j < n2
        if left[i] <= right[j]
            arr[k] = left[i]
            i = i + 1
        else
            arr[k] = right[j]
            j = j + 1
        k = k + 1

    // Copy any remaining elements of left[]
    while i < n1
        arr[k] = left[i]
        i = i + 1
        k = k + 1

    // Copy any remaining elements of right[]
    while j < n2
        arr[k] = right[j]
        j = j + 1
        k = k + 1
```

### Summary 

* **Divide**: Recursively split the array into two halves.
* **Conquer**: Recursively sort each half.
* **Combine**: Merge the two sorted halves using the `Merge()` function.
